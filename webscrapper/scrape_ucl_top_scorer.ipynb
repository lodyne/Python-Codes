```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.worldfootball.net/alltime_goalgetter/champions-league/tore/1/"
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup.title.text
```

```python
table = soup.find("table", class_="standard_tabelle")
```

```python
table
```

```python
rows = table.find_all("tr")
```

```python
rows
```

```python
data = []
for row in rows:
    item = []
    cells = row.find_all("td")
    for cell in cells:
        player = cell.text.strip()
        item.append(player)
    data.append(item)
```

```python
df = pd.DataFrame(data,columns=['#', 'Name', 'Clubs','Matches','Goals','Penalty','Goal per Match'])
```

```python
df
```

```python
drop_none_values = df.dropna()
```

```python
df_cleaned = drop_none_values.replace("\n", "-", regex=True)  # Replace all occurrences of \n with empty string
```

```python
df_cleaned = df_cleaned.map(lambda x: x.strip() if isinstance(x, str) else x)  # Apply strip to all string columns
```
