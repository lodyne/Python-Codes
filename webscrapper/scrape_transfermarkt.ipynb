```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# url = "https://www.worldfootball.net/teams/manchester-united/2024/2/"
# url = "https://www.capology.com/club/real-madrid/salaries/"
url = "https://www.transfermarkt.com/manchester-united/kader/verein/985"
```

```python
headers = {'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36 OPR/108.0.0.0"}
```

```python
response = requests.get(url,headers=headers)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup.title.text
```

```python
squad=[]
```

```python
table = soup.find("table", class_="items")
```

```python
table
```

```python
rows = table.find_all("tr") 
```

```python
rows
```

```python
data = []
for row in rows:
    item = []
    cells = row.find_all("td")
    for cell in cells:
        player = cell.text.strip()
        item.append(player)
    data.append(item)
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
drop_none_values = df.dropna()
```

```python
df_cleaned = drop_none_values.replace("\n", "", regex=True)  # Replace all occurrences of \n with empty string
```

```python
df_cleaned = df_cleaned.map(lambda x: x.strip() if isinstance(x, str) else x)  # Apply strip to all string columns
```

```python
df_cleaned
```

```python
df_cleaned.drop(df_cleaned.columns[[1,2,6]], axis=1, inplace=True)
```
